```python
import pandas as pd

# Reading the file
data = pd.read_csv('D:/Projects/Module 3/m3_h1/M3_H1/2017_jun_final.csv')
data
```

```python
#Step 5: Calculate Percentage of Blanks
missing_percentage = data.isnull().sum() / len(data) * 100
missing_percentage
```

```python
# Step 1: Remove all columns with any NaN values except the "Язык.программирования" column
columns_to_remove = data.columns[data.isnull().any() & (data.columns != 'Язык.программирования')]
data = data.drop(columns=columns_to_remove)
```

```python
# Step 2: Calculate the percentage of missing values again
missing_percentage = data.isnull().sum() / len(data) * 100
print(missing_percentage)
```

```python
# Create a new DataFrame with only Python specialists
python_data = data[data['Язык.программирования'] == 'Python']
python_data.shape
```

```python
# Group by the "Посада" column and aggregate data
grouped_data = python_data.groupby('Должность')['Зарплата.в.месяц'].agg(['min', 'max'])

# Create a new DataFrame from the aggregated data
aggregated_data = pd.DataFrame(grouped_data)

# Display the new DataFrame
aggregated_data
```

```python
# Define the function fill_avg_salary
def fill_avg_salary(row):
    return (row['min'] + row['max']) / 2

# Apply the function to each row to create a new column 'avg'
aggregated_data['avg'] = aggregated_data.apply(fill_avg_salary, axis=1)
aggregated_data
```

```python
# Generate descriptive statistics for the 'avg' column
descriptive_stats = aggregated_data['avg'].describe()
descriptive_stats
```

```python
# Save the DataFrame to a CSV file
file_path = 'D:/Projects/Module 3/m3_h1/M3_H1/aggregated_data.csv'  # Replace with your desired file path
aggregated_data.to_csv(file_path, index=False)
```

```python
import matplotlib.pyplot as plt

# Directive for Jupyter Notebook to build the graphs inline
%matplotlib inline

plt.figure(figsize=(10, 6))
aggregated_data['avg'].plot(kind='bar')
plt.title('Average Salaries for Each Position')
plt.xlabel('Position')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
aggregated_data['avg'].plot(kind='hist', bins=20, edgecolor='k')
plt.title('Distribution of Average Salaries')
plt.xlabel('Average Salary')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
aggregated_data[['min', 'max', 'avg']].plot(kind='box')
plt.title('Salary Range and Distribution')
plt.ylabel('Salary')
plt.show()
```
